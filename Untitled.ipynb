```python
import unicodedata

def normalize_basic(text):
    text = unicodedata.normalize("NFKC", text)  # unicode standard
    text = text.replace("\n", " ").replace("\r", " ").replace("\t", " ")
    text = text.lower()
    text = " ".join(text.split())  # collapse spaces
    return text
```

```python
import string

def remove_ponctuation(texte: str) -> str:
    table = str.maketrans("", "", string.punctuation)
    return texte.translate(table)
```

## Ponctuation

```python
import re

def remove_all_special_characters(text: str) -> str:
    # Keep letters (ASCII + accented), digits, and spaces only
    text = re.sub(r"[^A-Za-z0-9Ã€-Ã–Ã˜-Ã¶Ã¸-Ã¿\s]", "", text)
    # Remove extra spaces
    return re.sub(r"\s+", " ", text).strip()
```

## Stop word

```python
import nltk
from nltk.corpus import stopwords

def remove_stopwords(text: str) -> str:
    stop_words = set(stopwords.words("english"))
    words = text.split()
    filtered = [w for w in words if w.lower() not in stop_words]
    return " ".join(filtered)
```

## Number

```python
import re

def remove_number (texte: str) -> str:
    return re.sub(r"\d+", "", texte)
```

```python
import re

def remove_noise(text):
    text = re.sub(r"http\S+|www\S+", " URL ", text)
    text = re.sub(r"<.*?>", " ", text)
    text = re.sub(r"@\w+", " USER ", text)
    text = re.sub(r"#\w+", " ", text)
    return text
```

```python
def normalize_repetitions(text):
    return re.sub(r"(.)\1{2,}", r"\1\1", text)
```

## Insultes obfusquÃ©es (cachÃ©)

```python
def normalize_obfuscated_words(text):
    mapping = {
        "0": "o", "1": "i", "3": "e", "4": "a", "@": "a",
        "$": "s", "!": "i", "7": "t"
    }
    for k, v in mapping.items():
        text = text.replace(k, v)
    return text
```

```python
def clean_punctuation(text):
    text = re.sub(r"[!?]{2,}", " ! ", text)
    text = re.sub(r"\.{2,}", " . ", text)
    text = re.sub(r"[^a-zA-Z0-9'\s]", " ", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()
```

```python
data.drop_duplicates(subset="comment_text", inplace=True)
data = data[data["comment_text"].str.strip() != ""]
data = data[data["comment_text"].str.split().apply(len) > 2]
```

```python
def clean_comment(text):
    text = str(text)
    
    # 1. normalisation unicode + lowercase + trim
    text = normalize_basic(text)
    
    # 2. remove URLs, HTML, mentions
    text = remove_noise(text)

    text = remove_all_special_characters(text)
    
    text = remove_ponctuation(text)
    
    text = remove_stopwords(text)
    
    text = remove_number(text)

    # 3. emojis â†’ token EMOJI
    text = handle_emojis(text)
    
    # 4. reduit les rÃ©pÃ©titions artificielles
    text = normalize_repetitions(text)
    
    # 5. decode insultes obfusquÃ©es
    text = normalize_obfuscated_words(text)
    
    # 6. clean ponctuation excessive
    text = clean_punctuation(text)
    
    return text
```

```python
data["clean_text"] = data["comment_text"].apply(clean_comment)
```

## Transformation en vecteur & SMOTE

```python
# === VERSION ROBUSTE AVEC GESTION D'ERREURS ===

import pandas as pd
import numpy as np
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings('ignore')

def apply_smote_to_all_classes(data, target_columns):
    """
    Applique SMOTE Ã  toutes les classes cibles
    """
    # Vectorisation
    print("=== VECTORISATION ===")
    vectorizer = TfidfVectorizer(
        max_features=5000,
        stop_words='english',
        ngram_range=(1, 2),
        min_df=2,
        max_df=0.95
    )
    
    X_tfidf = vectorizer.fit_transform(data['clean_text'])
    print(f" Textes vectorisÃ©s: {X_tfidf.shape}")
    
    # Application SMOTE
    print("\n=== SMOTE ===")
    results = {}
    
    for target in target_columns:
        try:
            print(f"\nTraitement: {target}")
            y = data[target]
            
            # VÃ©rification distribution
            counter_before = Counter(y)
            print(f"   Avant SMOTE: {dict(counter_before)}")
            
            # ParamÃ¨tres adaptatifs
            n_minority = counter_before[1]
            k_neighbors = max(1, min(5, n_minority - 1))
            
            # Application SMOTE
            smote = SMOTE(random_state=42, k_neighbors=k_neighbors)
            X_res, y_res = smote.fit_resample(X_tfidf, y)
            
            # RÃ©sultats
            counter_after = Counter(y_res)
            print(f"   AprÃ¨s SMOTE: {dict(counter_after)}")
            print(f"   k_neighbors: {k_neighbors}")
            
            # Stockage
            results[target] = {
                'X_resampled': X_res,
                'y_resampled': y_res,
                'before': counter_before,
                'after': counter_after
            }
            
        except Exception as e:
            print(f" Erreur avec {target}: {e}")
            continue
    
    print(f"\n SMOTE succes: {len(results)}/{len(target_columns)} classes traitÃ©es")
    return results, vectorizer

# Application
target_columns = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
smote_results, vectorizer = apply_smote_to_all_classes(data, target_columns)
```

```python
# === VISUALISATION DES RÃ‰SULTATS ===
import matplotlib.pyplot as plt

def plot_smote_results(smote_results):
    """
    Visualise les distributions avant/aprÃ¨s SMOTE
    """
    n_classes = len(smote_results)
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.ravel()
    
    for idx, (target, result) in enumerate(smote_results.items()):
        if idx >= 6:  # Maximum 6 graphiques
            break
            
        before = result['before']
        after = result['after']
        
        labels = ['Non-' + target, target]
        before_counts = [before[0], before[1]]
        after_counts = [after[0], after[1]]
        
        x = np.arange(len(labels))
        width = 0.35
        
        axes[idx].bar(x - width/2, before_counts, width, label='Avant', alpha=0.7, color='blue')
        axes[idx].bar(x + width/2, after_counts, width, label='AprÃ¨s', alpha=0.7, color='red')
        
        axes[idx].set_title(f'{target.upper()}', fontweight='bold')
        axes[idx].set_xticks(x)
        axes[idx].set_xticklabels(labels)
        axes[idx].legend()
        axes[idx].grid(axis='y', alpha=0.3)
    
    plt.suptitle('DISTRIBUTIONS AVANT/APRÃˆS SMOTE', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()

# Affichage des graphiques
if smote_results:
    plot_smote_results(smote_results)
```

```python
# === IMPLÃ‰MENTATION CORRIGÃ‰E POUR VALIDATION.CSV SANS COMMENTAIRES ===

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
import warnings
warnings.filterwarnings('ignore')

# Configuration des chemins
TRAIN_PATH = "/Users/hortisdoumakpe/Desktop/PROJET/deep_learning/data/train.csv"
TEST_PATH = "/Users/hortisdoumakpe/Desktop/PROJET/deep_learning/data/test.csv"
VALIDATION_PATH = "/Users/hortisdoumakpe/Desktop/PROJET/deep_learning/data/validation.csv"

def load_all_datasets():
    """
    Charge tous les datasets : train, test et validation
    """
    print("=== CHARGEMENT DE TOUS LES DATASETS ===")
    
    # Chargement des donnÃ©es d'entraÃ®nement
    train_data = pd.read_csv(TRAIN_PATH)
    print(f"âœ… DonnÃ©es d'entraÃ®nement chargÃ©es: {train_data.shape}")
    
    # Chargement des donnÃ©es de test (sans labels)
    test_data = pd.read_csv(TEST_PATH)
    print(f"âœ… DonnÃ©es de test chargÃ©es: {test_data.shape}")
    
    # Chargement des donnÃ©es de validation (seulement labels)
    validation_data = pd.read_csv(VALIDATION_PATH)
    print(f"âœ… DonnÃ©es de validation chargÃ©es: {validation_data.shape}")
    
    # Affichage des colonnes disponibles
    print(f"ðŸ“Š Colonnes train: {list(train_data.columns)}")
    print(f"ðŸ“Š Colonnes test: {list(test_data.columns)}")
    print(f"ðŸ“Š Colonnes validation: {list(validation_data.columns)}")
    
    # Nettoyage des textes d'entraÃ®nement et test
    print("ðŸ§¹ Nettoyage des textes d'entraÃ®nement et test...")
    train_data['clean_text'] = train_data['comment_text'].apply(clean_comment)
    test_data['clean_text'] = test_data['comment_text'].apply(clean_comment)
    
    # PrÃ©paration des donnÃ©es de validation
    print("ðŸ”„ PrÃ©paration des donnÃ©es de validation...")
    
    # Joindre les commentaires du test.csv avec les labels du validation.csv
    validation_full = test_data.merge(validation_data, on='id', how='inner', suffixes=('', '_val'))
    
    print(f"âœ… DonnÃ©es de validation prÃ©parÃ©es: {validation_full.shape}")
    print(f"ðŸ“Š Colonnes validation aprÃ¨s jointure: {list(validation_full.columns)}")
    
    # VÃ©rification des valeurs manquantes dans les labels
    target_columns = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
    
    for target in target_columns:
        if target in validation_full.columns:
            missing_count = validation_full[target].isna().sum()
            print(f"   {target}: {missing_count} valeurs manquantes")
    
    # Colonnes cibles
    return train_data, test_data, validation_full, target_columns

def preprocess_validation_labels(validation_data, target_columns):
    """
    PrÃ©traite les labels de validation (-1 â†’ NaN â†’ suppression)
    """
    print("\n=== PRÃ‰TRAITEMENT DES LABELS DE VALIDATION ===")
    
    # Copie pour Ã©viter les modifications inplace
    validation_processed = validation_data.copy()
    
    # Remplacer -1 par NaN
    for target in target_columns:
        if target in validation_processed.columns:
            validation_processed[target] = validation_processed[target].replace(-1, np.nan)
    
    # Compter les NaN par colonne avant nettoyage
    print("ðŸ“Š Valeurs manquantes par colonne (aprÃ¨s remplacement -1 â†’ NaN):")
    for target in target_columns:
        if target in validation_processed.columns:
            nan_count = validation_processed[target].isna().sum()
            total_count = len(validation_processed)
            print(f"   {target}: {nan_count}/{total_count} ({nan_count/total_count*100:.1f}%)")
    
    # Supprimer les lignes oÃ¹ TOUTES les cibles sont NaN
    mask_all_nan = validation_processed[target_columns].isna().all(axis=1)
    validation_cleaned = validation_processed[~mask_all_nan].copy()
    
    print(f"âœ… Lignes supprimÃ©es (tous labels manquants): {mask_all_nan.sum()}")
    print(f"âœ… DonnÃ©es de validation aprÃ¨s nettoyage: {validation_cleaned.shape}")
    
    return validation_cleaned

def train_final_models(smote_results, vectorizer, target_columns):
    """
    EntraÃ®ne les modÃ¨les finaux sur toutes les donnÃ©es d'entraÃ®nement
    """
    print("\n=== ENTRAÃŽNEMENT DES MODÃˆLES FINAUX ===")
    
    final_models = {}
    
    for target_name in target_columns:
        print(f"\nðŸŽ¯ EntraÃ®nement final pour: {target_name}")
        
        if target_name not in smote_results:
            print(f"   âŒ DonnÃ©es SMOTE manquantes pour {target_name}")
            continue
        
        # Utilisation de toutes les donnÃ©es SMOTE pour l'entraÃ®nement final
        X_resampled = smote_results[target_name]['X_resampled']
        y_resampled = smote_results[target_name]['y_resampled']
        
        print(f"   DonnÃ©es d'entraÃ®nement SMOTE: {X_resampled.shape}")
        
        # Initialisation des modÃ¨les avec hyperparamÃ¨tres optimisÃ©s
        models = {
            'Logistic Regression': LogisticRegression(
                random_state=42, 
                max_iter=1000, 
                class_weight='balanced',
                C=1.0
            ),
            'Naive Bayes': MultinomialNB(alpha=0.1),
            'SVM': LinearSVC(
                random_state=42, 
                max_iter=1000, 
                class_weight='balanced',
                C=0.1
            ),
            'Random Forest': RandomForestClassifier(
                random_state=42, 
                n_estimators=100, 
                class_weight='balanced',
                max_depth=20
            )
        }
        
        # Ã‰valuation rapide pour sÃ©lectionner le meilleur modÃ¨le
        best_model_name = None
        best_model = None
        best_f1 = 0
        
        for model_name, model in models.items():
            try:
                print(f"   ðŸ”„ Test de: {model_name}")
                model.fit(X_resampled, y_resampled)
                
                # Validation croisÃ©e rapide
                from sklearn.model_selection import cross_val_score
                f1_scores = cross_val_score(model, X_resampled, y_resampled, 
                                          cv=3, scoring='f1', n_jobs=-1)
                mean_f1 = np.mean(f1_scores)
                
                print(f"      F1-score (CV): {mean_f1:.4f}")
                
                if mean_f1 > best_f1:
                    best_f1 = mean_f1
                    best_model_name = model_name
                    best_model = model
                    
            except Exception as e:
                print(f"      âŒ Erreur avec {model_name}: {e}")
                continue
        
        if best_model is not None:
            # EntraÃ®nement final sur toutes les donnÃ©es
            best_model.fit(X_resampled, y_resampled)
            final_models[target_name] = {
                'model_name': best_model_name,
                'model': best_model,
                'cv_f1_score': best_f1
            }
            print(f"   ðŸ† Meilleur modÃ¨le sÃ©lectionnÃ©: {best_model_name} (F1 CV: {best_f1:.4f})")
    
    return final_models

def evaluate_on_validation_set(final_models, vectorizer, validation_data, target_columns):
    """
    Ã‰valuation complÃ¨te sur le vrai jeu de validation
    """
    print("\n" + "="*60)
    print("ðŸ§ª Ã‰VALUATION SUR LE VRAI VALIDATION SET")
    print("="*60)
    
    validation_results = {}
    summary_results = []
    
    for target_name in target_columns:
        print(f"\nðŸŽ¯ Ã‰valuation pour: {target_name}")
        
        if target_name not in final_models:
            print(f"   âŒ ModÃ¨le manquant pour {target_name}")
            continue
        
        # PrÃ©parer les donnÃ©es pour cette cible spÃ©cifique
        target_data = validation_data[['id', 'clean_text', target_name]].dropna(subset=[target_name])
        
        if len(target_data) == 0:
            print(f"   âš ï¸  Aucune donnÃ©e valide pour {target_name}")
            continue
        
        # Vectorisation des donnÃ©es de validation
        X_val = vectorizer.transform(target_data['clean_text'])
        y_true = target_data[target_name]
        
        model_info = final_models[target_name]
        model = model_info['model']
        
        print(f"   Ã‰chantillons de validation: {len(target_data)}")
        print(f"   Distribution - Classe 0: {sum(y_true == 0)}, Classe 1: {sum(y_true == 1)}")
        
        # PrÃ©dictions
        y_pred = model.predict(X_val)
        
        # ProbabilitÃ©s pour AUC-ROC
        y_pred_proba = None
        if hasattr(model, 'predict_proba'):
            y_pred_proba = model.predict_proba(X_val)[:, 1]
        elif hasattr(model, 'decision_function'):
            y_pred_proba = model.decision_function(X_val)
        
        # MÃ©triques de performance
        accuracy = accuracy_score(y_true, y_pred)
        f1 = f1_score(y_true, y_pred, average='binary', zero_division=0)
        precision = precision_score(y_true, y_pred, average='binary', zero_division=0)
        recall = recall_score(y_true, y_pred, average='binary', zero_division=0)
        
        # AUC-ROC
        auc_roc = None
        if y_pred_proba is not None:
            try:
                auc_roc = roc_auc_score(y_true, y_pred_proba)
            except:
                auc_roc = None
        
        # Rapport de classification
        report = classification_report(y_true, y_pred, output_dict=True, zero_division=0)
        
        # Stockage des rÃ©sultats
        validation_results[target_name] = {
            'true_labels': y_true,
            'predictions': y_pred,
            'probabilities': y_pred_proba,
            'accuracy': accuracy,
            'f1_score': f1,
            'precision': precision,
            'recall': recall,
            'auc_roc': auc_roc,
            'classification_report': report,
            'sample_count': len(target_data)
        }
        
        summary_results.append({
            'Target': target_name,
            'Model': model_info['model_name'],
            'Samples': len(target_data),
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1-Score': f1,
            'AUC-ROC': auc_roc if auc_roc else 'N/A',
            'CV_F1_Score': model_info['cv_f1_score']
        })
        
        print(f"   âœ… Accuracy: {accuracy:.4f}")
        print(f"   âœ… F1-Score: {f1:.4f}")
        print(f"   âœ… Precision: {precision:.4f}")
        print(f"   âœ… Recall: {recall:.4f}")
        if auc_roc:
            print(f"   âœ… AUC-ROC: {auc_roc:.4f}")
    
    # DataFrame rÃ©capitulatif
    summary_df = pd.DataFrame(summary_results)
    
    print("\n" + "="*70)
    print("ðŸ“ˆ PERFORMANCES SUR VALIDATION SET (DONNÃ‰ES INVISIBLES)")
    print("="*70)
    print(summary_df.to_string(index=False))
    
    return validation_results, summary_df

def create_validation_visualizations(summary_df, validation_data, target_columns):
    """
    CrÃ©e des visualisations basÃ©es sur l'Ã©valuation du validation set
    """
    fig, axes = plt.subplots(2, 3, figsize=(20, 12))
    fig.suptitle('Performances RÃ©elles sur Validation Set (DonnÃ©es Invisibles)', 
                 fontsize=16, fontweight='bold')
    
    # 1. F1-Score par catÃ©gorie (mÃ©trique principale)
    if not summary_df.empty:
        axes[0, 0].barh(summary_df['Target'], summary_df['F1-Score'], color='lightcoral', alpha=0.7)
        axes[0, 0].set_title('F1-Score par CatÃ©gorie', fontsize=12)
        axes[0, 0].set_xlabel('F1-Score')
        axes[0, 0].set_xlim(0, 1)
        for i, v in enumerate(summary_df['F1-Score']):
            axes[0, 0].text(v + 0.01, i, f'{v:.3f}', va='center', fontsize=9)
    
    # 2. Nombre d'Ã©chantillons par catÃ©gorie
    sample_counts = []
    for target in target_columns:
        if target in validation_data.columns:
            valid_samples = validation_data[target].notna().sum()
            sample_counts.append({'Target': target, 'Samples': valid_samples})
    
    if sample_counts:
        sample_df = pd.DataFrame(sample_counts)
        axes[0, 1].barh(sample_df['Target'], sample_df['Samples'], color='lightblue', alpha=0.7)
        axes[0, 1].set_title('Ã‰chantillons Valides par CatÃ©gorie', fontsize=12)
        axes[0, 1].set_xlabel('Nombre d\'Ã©chantillons')
        for i, v in enumerate(sample_df['Samples']):
            axes[0, 1].text(v + 5, i, f'{v}', va='center', fontsize=9)
    
    # 3. Comparaison F1-Score CV vs Validation
    if not summary_df.empty:
        x_pos = range(len(summary_df))
        width = 0.35
        axes[0, 2].bar(x_pos, summary_df['CV_F1_Score'], width, label='F1 CV (Train)', alpha=0.7)
        axes[0, 2].bar([i + width for i in x_pos], summary_df['F1-Score'], width, label='F1 Validation', alpha=0.7)
        axes[0, 2].set_title('Comparaison F1-Score: CV vs Validation', fontsize=12)
        axes[0, 2].set_xlabel('CatÃ©gories')
        axes[0, 2].set_ylabel('F1-Score')
        axes[0, 2].set_xticks([i + width/2 for i in x_pos])
        axes[0, 2].set_xticklabels(summary_df['Target'], rotation=45, fontsize=9)
        axes[0, 2].legend()
        axes[0, 2].set_ylim(0, 1)
    
    # 4. Distribution des classes dans le validation set
    class_distribution = []
    for target in target_columns:
        if target in validation_data.columns:
            target_data = validation_data[target].dropna()
            if len(target_data) > 0:
                dist = target_data.value_counts()
                class_distribution.append({
                    'Target': target,
                    'Non-Toxique': dist.get(0, 0),
                    'Toxique': dist.get(1, 0)
                })
    
    if class_distribution:
        dist_df = pd.DataFrame(class_distribution)
        x = np.arange(len(dist_df))
        width = 0.35
        axes[1, 0].bar(x - width/2, dist_df['Non-Toxique'], width, label='Non-Toxique', alpha=0.7, color='gray')
        axes[1, 0].bar(x + width/2, dist_df['Toxique'], width, label='Toxique', alpha=0.7, color='red')
        axes[1, 0].set_title('Distribution des Classes (Validation Set)', fontsize=12)
        axes[1, 0].set_xlabel('CatÃ©gories')
        axes[1, 0].set_ylabel('Nombre d\'Ã©chantillons')
        axes[1, 0].set_xticks(x)
        axes[1, 0].set_xticklabels(dist_df['Target'], rotation=45, fontsize=9)
        axes[1, 0].legend()
    
    # 5. Accuracy par catÃ©gorie
    if not summary_df.empty:
        axes[1, 1].barh(summary_df['Target'], summary_df['Accuracy'], color='lightgreen', alpha=0.7)
        axes[1, 1].set_title('Accuracy par CatÃ©gorie', fontsize=12)
        axes[1, 1].set_xlabel('Accuracy')
        axes[1, 1].set_xlim(0, 1)
        for i, v in enumerate(summary_df['Accuracy']):
            axes[1, 1].text(v + 0.01, i, f'{v:.3f}', va='center', fontsize=9)
    
    # 6. Distribution des modÃ¨les sÃ©lectionnÃ©s
    if not summary_df.empty:
        model_counts = summary_df['Model'].value_counts()
        colors = plt.cm.Set3(np.linspace(0, 1, len(model_counts)))
        axes[1, 2].pie(model_counts.values, labels=model_counts.index, autopct='%1.1f%%', 
                       startangle=90, colors=colors)
        axes[1, 2].set_title('Distribution des ModÃ¨les SÃ©lectionnÃ©s', fontsize=12)
    
    plt.tight_layout()
    plt.show()

def generate_final_test_predictions(final_models, vectorizer, test_data):
    """
    GÃ©nÃ¨re les prÃ©dictions finales pour le test set
    """
    print("\n" + "="*50)
    print("ðŸ”® GÃ‰NÃ‰RATION DES PRÃ‰DICTIONS FINALES SUR TEST SET")
    print("="*50)
    
    # Vectorisation du test set
    X_test = vectorizer.transform(test_data['clean_text'])
    print(f"âœ… Test set vectorisÃ©: {X_test.shape}")
    
    # CrÃ©ation du DataFrame de rÃ©sultats
    results_df = test_data[['id', 'comment_text']].copy()
    
    # PrÃ©dictions pour chaque catÃ©gorie
    for target_name, model_info in final_models.items():
        model = model_info['model']
        
        print(f"ðŸŽ¯ PrÃ©dictions pour: {target_name}")
        
        # PrÃ©dictions binaires
        predictions = model.predict(X_test)
        results_df[target_name] = predictions.astype(int)
        
        # ProbabilitÃ©s si disponibles
        if hasattr(model, 'predict_proba'):
            probabilities = model.predict_proba(X_test)[:, 1]
            results_df[target_name + '_probability'] = probabilities
            print(f"   âœ… PrÃ©dictions + probabilitÃ©s gÃ©nÃ©rÃ©es")
        else:
            print(f"   âœ… PrÃ©dictions gÃ©nÃ©rÃ©es")
    
    print(f"\nâœ… PrÃ©dictions gÃ©nÃ©rÃ©es pour {len(results_df)} commentaires de test")
    
    return results_df

def save_final_results(final_models, vectorizer, validation_summary, test_predictions, validation_results):
    """
    Sauvegarde tous les rÃ©sultats finaux
    """
    print("\n" + "="*50)
    print("ðŸ’¾ SAUVEGARDE DES RÃ‰SULTATS FINAUX")
    print("="*50)
    
    # 1. Performances sur validation set
    if not validation_summary.empty:
        validation_summary.to_csv('final_validation_performance.csv', index=False)
        print("âœ… final_validation_performance.csv sauvegardÃ©")
    
    # 2. PrÃ©dictions pour le test set (format soumission)
    submission_cols = ['id', 'toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
    available_cols = [col for col in submission_cols if col in test_predictions.columns]
    submission_df = test_predictions[available_cols]
    submission_df.to_csv('final_submission.csv', index=False)
    print("âœ… final_submission.csv sauvegardÃ©")
    
    # 3. PrÃ©dictions dÃ©taillÃ©es avec probabilitÃ©s
    test_predictions.to_csv('final_test_predictions_detailed.csv', index=False)
    print("âœ… final_test_predictions_detailed.csv sauvegardÃ©")
    
    # 4. ModÃ¨les finaux
    import joblib
    models_dict = {target: info['model'] for target, info in final_models.items()}
    joblib.dump(models_dict, 'final_toxicity_models.pkl')
    joblib.dump(vectorizer, 'final_tfidf_vectorizer.pkl')
    print("âœ… ModÃ¨les et vectoriseur finaux sauvegardÃ©s")
    
    # 5. RÃ©sultats dÃ©taillÃ©s de validation
    if validation_results:
        validation_details = []
        for target_name, results in validation_results.items():
            validation_details.append({
                'Target': target_name,
                'Accuracy': results['accuracy'],
                'F1_Score': results['f1_score'],
                'Precision': results['precision'],
                'Recall': results['recall'],
                'AUC_ROC': results.get('auc_roc', 'N/A'),
                'True_Positives': sum((results['true_labels'] == 1) & (results['predictions'] == 1)),
                'False_Positives': sum((results['true_labels'] == 0) & (results['predictions'] == 1)),
                'True_Negatives': sum((results['true_labels'] == 0) & (results['predictions'] == 0)),
                'False_Negatives': sum((results['true_labels'] == 1) & (results['predictions'] == 0)),
                'Sample_Count': results['sample_count']
            })
        
        validation_details_df = pd.DataFrame(validation_details)
        validation_details_df.to_csv('validation_detailed_metrics.csv', index=False)
        print("âœ… validation_detailed_metrics.csv sauvegardÃ©")





```

```python
# === IMPLÃ‰MENTATION DE MODÃˆLES DEEP LEARNING POUR LA DÃ‰TECTION DE TOXICITÃ‰ ===

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import AutoTokenizer, AutoModel, get_linear_schedule_with_warmup
from torch.optim import AdamW
from tqdm.auto import tqdm
import math
import warnings
warnings.filterwarnings('ignore')

# Configuration des chemins
TRAIN_PATH = "/Users/hortisdoumakpe/Desktop/PROJET/deep_learning/data/train.csv"
TEST_PATH = "/Users/hortisdoumakpe/Desktop/PROJET/deep_learning/data/test.csv"
VALIDATION_PATH = "/Users/hortisdoumakpe/Desktop/PROJET/deep_learning/data/validation.csv"

# Configuration GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"ðŸŽ¯ Device utilisÃ©: {device}")

class ToxicityDataset(Dataset):
    """Dataset personnalisÃ© pour la dÃ©tection de toxicitÃ©"""
    def __init__(self, texts, labels, tokenizer, max_length=128):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length
    
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = self.labels[idx]
        
        encoding = self.tokenizer(
            text,
            truncation=True,
            padding='max_length',
            max_length=self.max_length,
            return_tensors='pt'
        )
        
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.float)
        }

class BERTToxicityClassifier(nn.Module):
    """Classifieur BERT pour dÃ©tection de toxicitÃ© multi-labels"""
    def __init__(self, n_classes, model_name='bert-base-uncased', dropout_rate=0.3):
        super(BERTToxicityClassifier, self).__init__()
        self.bert = AutoModel.from_pretrained(model_name)
        self.dropout = nn.Dropout(dropout_rate)
        self.classifier = nn.Linear(self.bert.config.hidden_size, n_classes)
        
    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        pooled_output = outputs.pooler_output
        output = self.dropout(pooled_output)
        return self.classifier(output)

class DistilBERTToxicityClassifier(nn.Module):
    """Version plus lÃ©gÃ¨re avec DistilBERT"""
    def __init__(self, n_classes, model_name='distilbert-base-uncased', dropout_rate=0.3):
        super(DistilBERTToxicityClassifier, self).__init__()
        self.distilbert = AutoModel.from_pretrained(model_name)
        self.dropout = nn.Dropout(dropout_rate)
        self.classifier = nn.Linear(self.distilbert.config.hidden_size, n_classes)
        
    def forward(self, input_ids, attention_mask):
        outputs = self.distilbert(input_ids=input_ids, attention_mask=attention_mask)
        hidden_state = outputs.last_hidden_state
        pooled_output = hidden_state[:, 0]  # Utiliser le token [CLS]
        output = self.dropout(pooled_output)
        return self.classifier(output)

class CNNTextClassifier(nn.Module):
    """Classifieur CNN pour texte avec multiples tailles de filtres"""
    def __init__(self, vocab_size, embedding_dim, n_classes, num_filters=100, filter_sizes=[3,4,5], dropout=0.5):
        super(CNNTextClassifier, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.convs = nn.ModuleList([
            nn.Conv1d(embedding_dim, num_filters, kernel_size=fs)
            for fs in filter_sizes
        ])
        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(num_filters * len(filter_sizes), n_classes)
        
    def forward(self, x):
        x = self.embedding(x)  # [batch_size, seq_len, embedding_dim]
        x = x.permute(0, 2, 1)  # [batch_size, embedding_dim, seq_len]
        
        conv_outputs = []
        for conv in self.convs:
            conv_out = torch.relu(conv(x))
            pooled = torch.max(conv_out, 2)[0]  # Max pooling sur la sÃ©quence
            conv_outputs.append(pooled)
            
        x = torch.cat(conv_outputs, dim=1)
        x = self.dropout(x)
        return self.fc(x)

class SimpleLSTMClassifier(nn.Module):
    """Classifieur LSTM simple pour comparaison"""
    def __init__(self, vocab_size, embedding_dim, hidden_dim, n_classes, n_layers=2, dropout=0.3):
        super(SimpleLSTMClassifier, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, n_layers, 
                           batch_first=True, dropout=dropout, bidirectional=True)
        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(hidden_dim * 2, n_classes)  # *2 pour bidirectional
        
    def forward(self, x):
        embedded = self.embedding(x)
        lstm_out, (hidden, cell) = self.lstm(embedded)
        
        # ConcatÃ©ner les derniers hidden states forward et backward
        hidden_forward = hidden[-2, :, :]  # Avant-dernier layer (forward)
        hidden_backward = hidden[-1, :, :]  # Dernier layer (backward)
        hidden_combined = torch.cat((hidden_forward, hidden_backward), dim=1)
        
        output = self.dropout(hidden_combined)
        return self.fc(output)

class DeepLearningTrainer:
    """Classe pour entraÃ®ner et Ã©valuer les modÃ¨les Deep Learning"""
    def __init__(self, model, device, model_type='bert'):
        self.model = model.to(device)
        self.device = device
        self.model_type = model_type
        
    def train_epoch(self, dataloader, optimizer, scheduler, criterion):
        self.model.train()
        total_loss = 0
        all_preds = []
        all_labels = []
        
        progress_bar = tqdm(dataloader, desc="Training")
        for batch in progress_bar:
            if self.model_type == 'bert' or self.model_type == 'distilbert':
                input_ids = batch['input_ids'].to(self.device)
                attention_mask = batch['attention_mask'].to(self.device)
                labels = batch['labels'].to(self.device)
                
                optimizer.zero_grad()
                outputs = self.model(input_ids=input_ids, attention_mask=attention_mask)
            else:
                inputs = batch['input_ids'].to(self.device)
                labels = batch['labels'].to(self.device)
                
                optimizer.zero_grad()
                outputs = self.model(inputs)
            
            loss = criterion(outputs, labels)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)
            optimizer.step()
            if scheduler:
                scheduler.step()
                
            total_loss += loss.item()
            preds = (torch.sigmoid(outputs) > 0.5).cpu().detach().numpy()
            all_preds.extend(preds)
            all_labels.extend(labels.cpu().detach().numpy())
            
            progress_bar.set_postfix({'loss': f'{loss.item():.4f}'})
            
        return total_loss / len(dataloader), np.array(all_preds), np.array(all_labels)
    
    def evaluate(self, dataloader, criterion):
        self.model.eval()
        total_loss = 0
        all_preds = []
        all_labels = []
        all_probabilities = []
        
        with torch.no_grad():
            for batch in tqdm(dataloader, desc="Evaluation"):
                if self.model_type == 'bert' or self.model_type == 'distilbert':
                    input_ids = batch['input_ids'].to(self.device)
                    attention_mask = batch['attention_mask'].to(self.device)
                    labels = batch['labels'].to(self.device)
                    
                    outputs = self.model(input_ids=input_ids, attention_mask=attention_mask)
                else:
                    inputs = batch['input_ids'].to(self.device)
                    labels = batch['labels'].to(self.device)
                    
                    outputs = self.model(inputs)
                
                loss = criterion(outputs, labels)
                total_loss += loss.item()
                
                probabilities = torch.sigmoid(outputs).cpu().numpy()
                preds = (probabilities > 0.5).astype(int)
                
                all_preds.extend(preds)
                all_labels.extend(labels.cpu().numpy())
                all_probabilities.extend(probabilities)
        
        return (total_loss / len(dataloader), 
                np.array(all_preds), 
                np.array(all_labels), 
                np.array(all_probabilities))

def prepare_deep_learning_data(train_data, validation_data, target_columns, tokenizer, batch_size=16):
    """PrÃ©pare les donnÃ©es pour l'entraÃ®nement Deep Learning"""
    print("ðŸ”„ PrÃ©paration des donnÃ©es pour Deep Learning...")
    
    # PrÃ©parer les donnÃ©es d'entraÃ®nement
    train_texts = train_data['clean_text'].tolist()
    train_labels = train_data[target_columns].values
    
    # PrÃ©parer les donnÃ©es de validation
    val_texts = []
    val_labels = []
    
    for target in target_columns:
        target_data = validation_data[['clean_text', target]].dropna()
        val_texts.extend(target_data['clean_text'].tolist())
        # CrÃ©er des labels multi-hot encoding
        multi_labels = np.zeros((len(target_data), len(target_columns)))
        for i, (idx, row) in enumerate(target_data.iterrows()):
            multi_labels[i] = [row[col] if col in target_data.columns else 0 for col in target_columns]
        val_labels.extend(multi_labels)
    
    # CrÃ©er les datasets
    train_dataset = ToxicityDataset(train_texts, train_labels, tokenizer)
    val_dataset = ToxicityDataset(val_texts, np.array(val_labels), tokenizer)
    
    # CrÃ©er les dataloaders
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    
    print(f"âœ… DonnÃ©es prÃ©parÃ©es - Train: {len(train_dataset)}, Validation: {len(val_dataset)}")
    
    return train_loader, val_loader

def train_deep_learning_models(train_data, validation_data, target_columns):
    """EntraÃ®ne plusieurs modÃ¨les Deep Learning"""
    print("ðŸš€ ENTRAÃŽNEMENT DES MODÃˆLES DEEP LEARNING")
    print("="*50)
    
    # Configuration
    n_classes = len(target_columns)
    batch_size = 16
    num_epochs = 2  # RÃ©duit pour les tests
    learning_rate = 2e-5
    
    # Initialisation du tokenizer
    tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
    
    # PrÃ©paration des donnÃ©es
    train_loader, val_loader = prepare_deep_learning_data(
        train_data, validation_data, target_columns, tokenizer, batch_size
    )
    
    # ModÃ¨les Ã  entraÃ®ner (version simplifiÃ©e)
    models_config = {
        'BERT': {
            'model': BERTToxicityClassifier(n_classes=n_classes),
            'type': 'bert',
            'lr': 2e-5
        },
        'DistilBERT': {
            'model': DistilBERTToxicityClassifier(n_classes=n_classes),
            'type': 'distilbert', 
            'lr': 2e-5
        }
    }
    
    trained_models = {}
    results = {}
    
    for model_name, config in models_config.items():
        print(f"\nðŸŽ¯ EntraÃ®nement du modÃ¨le: {model_name}")
        print("="*40)
        
        model = config['model']
        model_type = config['type']
        lr = config['lr']
        
        # Initialisation du trainer
        trainer = DeepLearningTrainer(model, device, model_type)
        
        # Optimizer et scheduler
        optimizer = AdamW(model.parameters(), lr=lr, weight_decay=0.01)
        total_steps = len(train_loader) * num_epochs
        scheduler = get_linear_schedule_with_warmup(
            optimizer, num_warmup_steps=0, num_training_steps=total_steps
        )
        
        # Loss function (Multi-label binary classification)
        criterion = nn.BCEWithLogitsLoss()
        
        # EntraÃ®nement
        best_val_f1 = 0
        train_losses = []
        val_losses = []
        
        for epoch in range(num_epochs):
            print(f"\nðŸ“š Epoch {epoch+1}/{num_epochs}")
            
            # EntraÃ®nement
            train_loss, train_preds, train_labels = trainer.train_epoch(
                train_loader, optimizer, scheduler, criterion
            )
            
            # Validation
            val_loss, val_preds, val_labels, val_probs = trainer.evaluate(val_loader, criterion)
            
            # Calcul des mÃ©triques
            train_f1 = f1_score(train_labels, train_preds, average='macro', zero_division=0)
            val_f1 = f1_score(val_labels, val_preds, average='macro', zero_division=0)
            
            train_losses.append(train_loss)
            val_losses.append(val_loss)
            
            print(f"   Train Loss: {train_loss:.4f}, Train F1: {train_f1:.4f}")
            print(f"   Val Loss: {val_loss:.4f}, Val F1: {val_f1:.4f}")
            
            # Sauvegarde du meilleur modÃ¨le
            if val_f1 > best_val_f1:
                best_val_f1 = val_f1
                torch.save(model.state_dict(), f'best_{model_name.lower()}_model.pth')
                print(f"   ðŸ’¾ Meilleur modÃ¨le sauvegardÃ© (F1: {val_f1:.4f})")
        
        # Stockage des rÃ©sultats
        trained_models[model_name] = {
            'model': model,
            'trainer': trainer,
            'tokenizer': tokenizer
        }
        
        results[model_name] = {
            'train_losses': train_losses,
            'val_losses': val_losses,
            'best_val_f1': best_val_f1,
            'val_predictions': val_preds,
            'val_probabilities': val_probs,
            'val_labels': val_labels
        }
        
        # Visualisation de l'apprentissage
        plot_training_curves(train_losses, val_losses, model_name)
    
    return trained_models, results

def plot_training_curves(train_losses, val_losses, model_name):
    """Visualise les courbes d'apprentissage"""
    plt.figure(figsize=(10, 6))
    plt.plot(train_losses, label='Train Loss', linewidth=2)
    plt.plot(val_losses, label='Validation Loss', linewidth=2)
    plt.title(f'Courbes d\'Apprentissage - {model_name}', fontsize=14)
    plt.xlabel('Epochs', fontsize=12)
    plt.ylabel('Loss', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(f'training_curves_{model_name}.png', dpi=300, bbox_inches='tight')
    plt.show()

def evaluate_deep_learning_models(trained_models, results, target_columns):
    """Ã‰valuation complÃ¨te des modÃ¨les Deep Learning"""
    print("\n" + "="*60)
    print("ðŸ“Š Ã‰VALUATION DES MODÃˆLES DEEP LEARNING")
    print("="*60)
    
    summary_results = []
    
    for model_name, result in results.items():
        print(f"\nðŸ” Ã‰valuation de {model_name}:")
        
        val_labels = result['val_labels']
        val_preds = result['val_predictions']
        val_probs = result['val_probabilities']
        
        # MÃ©triques globales
        global_accuracy = accuracy_score(val_labels, val_preds)
        global_f1 = f1_score(val_labels, val_preds, average='macro', zero_division=0)
        
        # MÃ©triques par catÃ©gorie
        category_metrics = []
        for i, target in enumerate(target_columns):
            if i < val_preds.shape[1]:  # VÃ©rifier que l'index existe
                accuracy = accuracy_score(val_labels[:, i], val_preds[:, i])
                f1 = f1_score(val_labels[:, i], val_preds[:, i], average='binary', zero_division=0)
                precision = precision_score(val_labels[:, i], val_preds[:, i], average='binary', zero_division=0)
                recall = recall_score(val_labels[:, i], val_preds[:, i], average='binary', zero_division=0)
                
                # AUC-ROC si disponible
                auc_roc = None
                if val_probs is not None and len(val_probs) > 0:
                    try:
                        auc_roc = roc_auc_score(val_labels[:, i], val_probs[:, i])
                    except:
                        auc_roc = None
                
                category_metrics.append({
                    'Target': target,
                    'Accuracy': accuracy,
                    'Precision': precision,
                    'Recall': recall,
                    'F1-Score': f1,
                    'AUC-ROC': auc_roc if auc_roc else 'N/A'
                })
                
                print(f"   {target}: F1={f1:.4f}, AUC={auc_roc if auc_roc else 'N/A'}")
        
        summary_results.append({
            'Model': model_name,
            'Global_Accuracy': global_accuracy,
            'Global_F1_Score': global_f1,
            'Best_Val_F1': result['best_val_f1'],
            'Category_Metrics': category_metrics
        })
        
        print(f"   Global - Accuracy: {global_accuracy:.4f}, F1: {global_f1:.4f}")
    
    return summary_results

def predict_with_deep_learning(trained_models, test_data, target_columns):
    """GÃ©nÃ¨re des prÃ©dictions avec les modÃ¨les Deep Learning"""
    print("\n" + "="*50)
    print("ðŸ”® PRÃ‰DICTIONS AVEC DEEP LEARNING")
    print("="*50)
    
    all_predictions = {}
    
    for model_name, model_info in trained_models.items():
        print(f"\nðŸŽ¯ PrÃ©dictions avec {model_name}...")
        
        model = model_info['model']
        tokenizer = model_info['tokenizer']
        
        # PrÃ©paration des donnÃ©es de test
        test_texts = test_data['clean_text'].tolist()
        test_dataset = ToxicityDataset(test_texts, np.zeros((len(test_texts), len(target_columns))), tokenizer)
        test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
        
        # PrÃ©dictions
        model.eval()
        all_preds = []
        all_probs = []
        
        with torch.no_grad():
            for batch in tqdm(test_loader, desc=f"Predicting with {model_name}"):
                if hasattr(model, 'bert') or hasattr(model, 'distilbert'):
                    input_ids = batch['input_ids'].to(device)
                    attention_mask = batch['attention_mask'].to(device)
                    outputs = model(input_ids=input_ids, attention_mask=attention_mask)
                else:
                    inputs = batch['input_ids'].to(device)
                    outputs = model(inputs)
                
                probabilities = torch.sigmoid(outputs)
                preds = (probabilities > 0.5).cpu().numpy()
                
                all_probs.extend(probabilities.cpu().numpy())
                all_preds.extend(preds)
        
        # CrÃ©ation du DataFrame de rÃ©sultats
        results_df = test_data[['id', 'comment_text']].copy()
        predictions_array = np.array(all_preds)
        probabilities_array = np.array(all_probs)
        
        for i, target in enumerate(target_columns):
            if i < predictions_array.shape[1]:
                results_df[target] = predictions_array[:, i].astype(int)
                results_df[f'{target}_probability'] = probabilities_array[:, i]
        
        all_predictions[model_name] = results_df
        print(f"âœ… {model_name}: {len(results_df)} prÃ©dictions gÃ©nÃ©rÃ©es")
    
    return all_predictions

def ensemble_predictions(all_predictions, target_columns):
    """Combine les prÃ©dictions de plusieurs modÃ¨les"""
    print(f"\nðŸ¤ COMBINAISON DES PRÃ‰DICTIONS (ENSEMBLE)")
    
    # Prendre le premier modÃ¨le comme base
    ensemble_df = all_predictions[list(all_predictions.keys())[0]][['id', 'comment_text']].copy()
    
    for target in target_columns:
        # Collecter les prÃ©dictions de tous les modÃ¨les
        model_preds = []
        model_probs = []
        
        for model_name, pred_df in all_predictions.items():
            if target in pred_df.columns:
                model_preds.append(pred_df[target].values)
                if f'{target}_probability' in pred_df.columns:
                    model_probs.append(pred_df[f'{target}_probability'].values)
        
        if model_preds:
            # Voting majoritaire
            stacked_preds = np.stack(model_preds, axis=1)
            ensemble_preds = (stacked_preds.sum(axis=1) > len(model_preds) / 2).astype(int)
            ensemble_df[target] = ensemble_preds
            
            # Moyenne des probabilitÃ©s
            if model_probs:
                prob_arrays = np.stack(model_probs, axis=1)
                ensemble_df[f'{target}_probability'] = np.mean(prob_arrays, axis=1)
    
    return ensemble_df

# === FONCTIONS UTILITAIRES POUR LE CHARGEMENT DES DONNÃ‰ES ===

def load_all_datasets():
    """Charge tous les datasets"""
    print("=== CHARGEMENT DES DONNÃ‰ES ===")
    
    train_data = pd.read_csv(TRAIN_PATH)
    test_data = pd.read_csv(TEST_PATH)
    validation_data = pd.read_csv(VALIDATION_PATH)
    
    print(f"âœ… Train: {train_data.shape}, Test: {test_data.shape}, Validation: {validation_data.shape}")
    
    # Nettoyage des textes
    train_data['clean_text'] = train_data['comment_text'].apply(clean_comment)
    test_data['clean_text'] = test_data['comment_text'].apply(clean_comment)
    
    # Jointure validation avec test pour rÃ©cupÃ©rer les commentaires
    validation_full = test_data.merge(validation_data, on='id', how='inner', suffixes=('', '_val'))
    validation_full['clean_text'] = validation_full['comment_text'].apply(clean_comment)
    
    target_columns = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
    
    return train_data, test_data, validation_full, target_columns

def preprocess_validation_labels(validation_data, target_columns):
    """PrÃ©traite les labels de validation"""
    validation_processed = validation_data.copy()
    
    for target in target_columns:
        if target in validation_processed.columns:
            validation_processed[target] = validation_processed[target].replace(-1, np.nan)
    
    return validation_processed

# === PIPELINE DEEP LEARNING COMPLET ===

print("ðŸš€ DÃ‰MARRAGE DU PIPELINE DEEP LEARNING")
print("="*60)

try:
    # 1. Chargement des donnÃ©es
    train_data, test_data, validation_data, target_columns = load_all_datasets()
    validation_processed = preprocess_validation_labels(validation_data, target_columns)
    
    # 2. EntraÃ®nement des modÃ¨les Deep Learning
    trained_models, training_results = train_deep_learning_models(
        train_data, validation_processed, target_columns
    )
    
    # 3. Ã‰valuation dÃ©taillÃ©e
    dl_summary = evaluate_deep_learning_models(trained_models, training_results, target_columns)
    
    # 4. PrÃ©dictions sur le test set
    dl_predictions = predict_with_deep_learning(trained_models, test_data, target_columns)
    
    # 5. Combinaison des prÃ©dictions (Ensemble)
    ensemble_predictions_df = ensemble_predictions(dl_predictions, target_columns)
    
    # 6. Sauvegarde des rÃ©sultats
    print("\nðŸ’¾ SAUVEGARDE DES RÃ‰SULTATS DEEP LEARNING")
    
    ensemble_predictions_df.to_csv('deep_learning_ensemble_predictions.csv', index=False)
    print("âœ… deep_learning_ensemble_predictions.csv sauvegardÃ©")
    
    for model_name, pred_df in dl_predictions.items():
        pred_df.to_csv(f'deep_learning_{model_name.lower()}_predictions.csv', index=False)
        print(f"âœ… deep_learning_{model_name.lower()}_predictions.csv sauvegardÃ©")
    
    # Rapport final
    print("\n" + "="*70)
    print("ðŸŽ‰ PIPELINE DEEP LEARNING TERMINÃ‰ AVEC SUCCÃˆS!")
    print("="*70)
    
    print(f"\nðŸ† PERFORMANCES DES MODÃˆLES DEEP LEARNING:")
    for result in dl_summary:
        print(f"   â€¢ {result['Model']}: F1={result['Global_F1_Score']:.4f}, Accuracy={result['Global_Accuracy']:.4f}")
    
    print(f"\nðŸ“Š DISTRIBUTION DES PRÃ‰DICTIONS ENSEMBLE:")
    for target in target_columns:
        if target in ensemble_predictions_df.columns:
            positive_count = ensemble_predictions_df[target].sum()
            positive_percent = (positive_count / len(ensemble_predictions_df)) * 100
            print(f"   â€¢ {target}: {positive_count} positifs ({positive_percent:.2f}%)")
    
    # Affichage d'un Ã©chantillon
    print("\nðŸ“‹ Ã‰CHANTILLON DES PRÃ‰DICTIONS ENSEMBLE:")
    sample_cols = ['id', 'toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
    available_cols = [col for col in sample_cols if col in ensemble_predictions_df.columns]
    sample_predictions = ensemble_predictions_df[available_cols].head(6)
    print(sample_predictions.to_string(index=False))

except Exception as e:
    print(f"âŒ Erreur lors de l'exÃ©cution: {e}")
    import traceback
    traceback.print_exc()
```
